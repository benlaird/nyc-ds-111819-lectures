```python
from itertools import combinations, permutations, combinations_with_replacement
```

```python
list(permutations(range(10)))
```

```python
list(permutations(['hi', 'bye', 'sigh'])
```

```python
combinations(set('abcdefgh'), 5)
```

```python
len(list(permutations(set('abcdefgh'), 5)))
```

```python
len(list(combinations_with_replacement(set('abcdefgh'), 5)))
```
