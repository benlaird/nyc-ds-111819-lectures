### Let's try approximating the normal distribution by sampling from other types of distributions!

1. Draw samples from the binomial distribution to get an approximately normal distribution (play with the parameters until you get there!) Graph the histogram in seaborn and matplotlib.

2. Can you approximate the normal distribution by sampling from dice rolls in the same way? Try simulating multiple dice rolls (2, 3, 4) and plotting the results on a histogram.

3. Last let's try the same process with the poisson distribution (np.random.poisson)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline
```

```python
def experiment1 (n):
    results = []
    p= 0.5
    for i in range(n):
        test_result = np.random.binomial(n, p, size = 100)
        results.append(test_result.mean())
    return test_result
```

```python
# np.random.binomial(10, .5, size=10)
```

```python
plot_var = experiment1(100_000)
```

```python
x = plot_var
sns.distplot(x)
plt.show()
```

```python
def dice_rolls(n):
    results = []
    for i in range(n):
        test_result = np.random.randint(1, 7, size = 100)
        results.append(test_result.mean())
    return results
```

```python
dice_roll = dice_rolls(10_000)
```

```python

sns.distplot(dice_roll)
plt.show()
```

```python
def poisson(n):
    results = []
    for i in range(n):
        test_result = np.random.poisson(3, size = 100)
        results.append(test_result.mean())
    return results
```

```python
p = poisson(10_000)
# print(p)
sns.distplot(p)
plt.show()
```
